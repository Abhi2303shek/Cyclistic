```python
import pandas as pd
```

```python
df1 = pd.read_csv("202101-divvy-tripdata.csv")
df2 = pd.read_csv("202102-divvy-tripdata.csv")
df3 = pd.read_csv("202103-divvy-tripdata.csv")
```

```python
cyclist = pd.concat([df1, df2, df3], ignore_index=True)
cyclist
```

```python
cyclist['started_at'] = pd.to_datetime(cyclist['started_at'])
cyclist['ended_at'] = pd.to_datetime(cyclist['ended_at'])
```

```python
cyclist['start_date'] = cyclist['started_at'].dt.date
cyclist['end_date'] = cyclist['ended_at'].dt.date
```

```python
cyclist['start_time'] = cyclist['started_at'].dt.time
cyclist['end_time'] = cyclist['ended_at'].dt.time
```

```python
cyclist
```

```python
missing_percent = cyclist['end_station_name'].isna().mean() * 100
print(f"Missing end station %: {missing_percent:.2f}%")
```

```python
cyclist['end_station_name'] = cyclist['end_station_name'].fillna("Unknown / Dockless")
cyclist['end_station_id'] = cyclist['end_station_id'].fillna("Unknown")
```

```python
cyclist = cyclist.drop(columns=['started_at','ended_at'])
```

```python
cyclist
```

```python
cyclist.to_csv('cyclist.csv')
```
